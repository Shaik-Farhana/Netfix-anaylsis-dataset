```python
import pandas as pd
df = pd.read_csv('netflix_titles.csv')
df.head()
df.shape
df.info()
df.head()
df.isnull().sum()
```

```python
**Dataset Summary:**
- Rows: 8807
- Columns: 12
- Null values in: director, cast, rating, etc.
```

```python
# Exploring Netflix content Releases by Year
import seaborn as sns
import matplotlib.pyplot as plt

df['release_year'].value_counts().sort_index().plot(kind='bar', figsize=(12,6))
plt.title("Netflix Releases Over the Years")
plt.xlabel("Year")
plt.ylabel("Number of Releases")
plt.show()
```

```python
# Top 10 genre
df['listed_in'].value_counts().head(10).plot(kind='barh', color='purple')
```

```python
# Shows by country 
df['country'].value_counts().head(10).plot(kind='bar', color='lightblue')
```

```python
# This chart shows country wise release of Netflix content over the past decade, United States has the hightest number of releases 
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(data=df, x='type', palette='dark')
plt.title('TV Shows vs Movies')
```
